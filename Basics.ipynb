Now let's do some basic data exploration to determine what other cleaning 
tasks need to be done. Initially we will look for: - Text columns where 
all or almost all values are the same. These can often be dropped as they 
don't have useful information for analysis. - Examples of numeric data

Most likely, we shall drop nrofpictures (basicaly no values there) and datecrawled (equal to ad_created) columns. We shall need to make new digit columns, remove "$" symbol from price, "km" from odometer column

```python
#removing '$' from price and stripping of extra spaces
autos['price'] = (autos['price']
                  .str.replace("$","")
                  .str.replace(",","")
                  .str.strip()
                 )
#removing 'km' from odometer and stripping of extra spaces
autos['odometer'] = (autos['odometer']
                     .str.replace("km","")
                     .str.replace(",","")
                     .str.strip()
                    )
#renaming columns
autos.rename( {'odometer':'odometer_km'}, axis = 1, inplace = True)
#making both columns numeric type
autos['price'] = autos['price'].astype(int)
autos['odometer_km'] = autos['odometer_km'].astype(int)
```

```python
def description(series, tf = False):
    #this function gives overall description over digit columns
    print('unique values', series, "\n", autos[series].unique().shape, "\n")
    print(series, 'parameters', "\n", autos[series].describe(), "\n")
    print('value counts', series, "\n", autos[series].
      value_counts(dropna=False)
      .head()
      .sort_index(ascending = tf))
    return series

description('odometer_km', False)
description('price', True)   
autos["price"].value_counts().sort_index(ascending=False).head(20)
```

the next step in our research is data aggregation.
We shall:
- Identify the unique values we want to aggregate by
- Create an empty dictionary to store our aggregate data
- Loop over the unique values, and for each:
    - Subset the dataframe by the unique values
    - Calculate the mean of whichever column we're interested in
    - Assign the val/mean to the dict as k/v.

```python
autos['brand'].value_counts(normalize = True).head(20)
```

I'm aggregating for top 20 brends in dataset

```python
mean_price_by_brand = {}
```

```python
brands = autos['brand'].value_counts(normalize = True).head(20).index
brands
```

```python
for brand in brands:
    selected_brand = autos[autos['brand'] == brand]
    mean_price = selected_brand['price'].mean()
    mean_price_by_brand[brand] = int(mean_price)
mean_price_by_brand
```

most expencive brand sonstige_autos, least expensive - fiat

```python
top_6_brands = autos['brand'].value_counts(normalize = True).head(6).index
mean_price_d = {}
mean_mileage = {}
for brand in top_6_brands:
    selected_brand = autos[autos['brand'] == brand]
    mean_price = selected_brand['price'].mean()
    mean_price_d[brand] = int(mean_price)
for brand in top_6_brands:
    selected_brand = autos[autos['brand'] == brand]
    mean_odometer = selected_brand['odometer_km'].mean()
    mean_mileage[brand] = int(mean_odometer)
```

```python
top_6_brands = pd.Series(top_6_brands)
```

```python
df = pd.DataFrame(top_6_brands,columns =['mean_price_d'] )
df["mean_mileage"] = mean_mileage 
```
